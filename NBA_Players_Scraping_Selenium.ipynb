```python
# Loading the player list from the pickle file
with open("player_list.pkl","rb") as f:
    loaded_list = pickle.load(f)
```

```python
# Define a dataframe to insert the scraped data
nba_players_info_df = pd.DataFrame()
```

```python
for each_element in loaded_list[:20]:
    try:
        dict_to_append = {}
        dict_to_append['NAME'] = each_element
        browser.find_element_by_link_text(each_element).click()
        elements_alex_title = browser.find_elements_by_class_name('player-stats__stat-title')
        elements_alex_value = browser.find_elements_by_class_name('player-stats__stat-value')
        for e1,e2 in zip(elements_alex_title, elements_alex_value):
            if e2.text:
                dict_to_append[e1.text] = e2.text
            else:
                dict_to_append[e1.text] = None
        browser.execute_script("window.history.go(-1)")
        nba_players_info_df = nba_players_info_df.append(dict_to_append, ignore_index=True)
        
    
    except exceptions.StaleElementReferenceException:
         pass
```

```python
nba_players_info_df = nba_players_info_df.iloc[:, :-1]
nba_players_info_df
```

```python
nba_players_info_df.to_csv("nba_players_info.csv", index=False)
```
